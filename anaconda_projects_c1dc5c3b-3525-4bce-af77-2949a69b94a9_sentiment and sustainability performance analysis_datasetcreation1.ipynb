```python
brands = ["BMW", "Mercedes-Benz", "Volkswagen", "Audi", "Porsche"]
```

```python
years = list(range(2018, 2026))
```

```python
reviews_per_brand_year = 200
```

```python
# Sentiment components
```

```python
sentiments = ["Positive", "Neutral", "Negative"]
```

```python
sentiment_weights = [0.55, 0.25, 0.20]  # Slight bias toward positive
```

```python
# Synthetic review templates
```

```python
positive_templates = [
    "I love my {brand}, it feels modern and efficient!",
    "{brand} really nailed the electric design this year.",
    "Smooth ride and amazing performance from {brand}.",
    "Impressive range and eco-friendly features by {brand}.",
    "The {brand} is worth every penny — luxurious and sustainable."
]
```

```python
neutral_templates = [
    "The {brand} is okay, not much to complain about.",
    "My experience with {brand} has been average so far.",
    "{brand} offers decent performance but could improve charging speed.",
    "Not bad, but not outstanding either — just fine for daily use.",
    "Overall, {brand} delivers what you expect for the price."
]
```

```python
negative_templates = [
    "I'm disappointed with my {brand}, the battery drains too fast.",
    "{brand} needs to work on their EV reliability.",
    "Too expensive for what it offers — not impressed by {brand}.",
    "Charging time is too long and the range is poor.",
    "The {brand} interior feels outdated compared to competitors."
]
```

```python
import random # Add this import at the beginning of your code

for brand in brands:
    base_co2 = random.randint(130, 160)
    base_ev_share = random.uniform(5, 10)
    for year in years:
        # Adjust sustainability metrics over time
        co2 = base_co2 - (year - 2018) * random.uniform(3, 6)
        ev_share = base_ev_share + (year - 2018) * random.uniform(4, 7)
        sustainability_index = round((1 - co2 / 160) * 0.6 + (ev_share / 100) * 0.4, 3)

```

```python
for _ in range(reviews_per_brand_year):
            sentiment_label = random.choices(sentiments, weights=sentiment_weights)[0]
            if sentiment_label == "Positive":
                template = random.choice(positive_templates)
                sentiment_score = round(random.uniform(0.4, 1.0), 2)
            elif sentiment_label == "Neutral":
                template = random.choice(neutral_templates)
                sentiment_score = round(random.uniform(-0.2, 0.4), 2)
            else:
                template = random.choice(negative_templates)
                sentiment_score = round(random.uniform(-1.0, -0.2), 2)

            review_text = template.format(brand=brand)

            data.append([brand, year, review_text, sentiment_label, sentiment_score,
                         round(co2, 1), round(ev_share, 1), sustainability_index])
```

```python
# Create DataFrame
```

```python
df = pd.DataFrame(data, columns=[
    "brand", "year", "review_text", "sentiment_label", "sentiment_score",
    "avg_co2_emission_gkm", "ev_share_percent", "sustainability_index"
])

# Show preview
df.head(10)
```
